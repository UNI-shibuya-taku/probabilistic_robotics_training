```python
from scipy.stats import expon, norm,uniform # 指数分布の機能を提供するオブジェクト
import matplotlib.pyplot as plt
s = expon(scale = 1.0 / (1e-100 + 5))
t = s.rvs()
u = []
for i in range(100):
    u.append(s.rvs())
print(t)
plt.plot(u)
plt.show(u)
```
